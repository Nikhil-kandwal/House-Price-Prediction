it is important to ensure that your feature names do not contain any characters like [,] or <.
These characters are invalid in feature names for xgboost.

So remove this type of feature name

ensure all feature name are string 

```python
x_train.columns=x_train.columns.astype(str)
```

```python
x_train.columns= x_train.columns.str.replace(r'[<>[\]]','',regex=True)
```

```python
x_train
```

```python
xgb_model=xgb.XGBRegressor()
```

```python
xgb_model.fit(x_train,y_train)
```

using hyperparameter tuning

```python
from sklearn.model_selection import GridSearchCV
```

```python
xgb_model2=xgb.XGBRegressor()
```

```python
param_grid_eg = {
    'n_estimators': [100, 200, 300],  # number of trees
    'learning_rate': [0.01, 0.05, 0.1],  # step size
    'max_depth': [3, 5, 7],  # tree depth
    'min_child_weight': [1, 3, 5],  # minimum sum of instance weight
    'subsample': [0.8, 0.9, 1.0],  # fraction of samples to use
    'colsample_bytree': [0.8, 0.9, 1.0],  # fraction of features
}
```

```python
grid_search = GridSearchCV(estimator=xgb_model2, 
                           param_grid=param_grid_eg, 
                           scoring='r2',  # Evaluation metric
                           cv=5,  # 5-fold cross-validation
                           verbose=1, 
                           n_jobs=-1)  # Use all available cores
```

```python
grid_search.fit(x_train,y_train)
```

```python
best_params = grid_search.best_params_
```

```python
best_params
```

```python
best_model = grid_search.best_estimator_
```

```python
xgb_model_pred2=best_model.predict(x_test)
```

```python
r2_score(y_test,xgb_model_pred2)
```

```python
xgb_model_pred2=best_model.predict(x_train)
r2_score(y_train,xgb_model_pred2)
```
